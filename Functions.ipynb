1. What is the difference between a function and a method in Python?

-> Functions :-  
-> These are standalone blocks of code that can be called from anywhere in the program. Functions are used to perform a specific task, and they can be reused. Functions can be used in both object-oriented and non-object-oriented programming languages.   
-> Methods :-  
-> These are sets of instructions that are associated with an object or class. Methods are used to encapsulate behaviors within objects, and they adhere to the principles of object-oriented programming. Methods can only be used in object-oriented programming languages.

-> The  differences between a function and a method are:-   
  
*  Functions are independent blocks of code, while methods are associated with objects or classes
*  Functions are called by name, while methods are accessed using dot notation
*  The first argument of a method should be "self", but functions do not require this parameter.
*  Functions perform operations on data that the user passes as parameters or arguments. Methods have access to the data of the object they are associated with.

-> Here's an example to illustrate the difference:


  #function  
def greet(name):   
    print(f"Hello, {name}!")

  #Class with method

    def __init__(self, name):  
        self.name = name

 def greet(self):  
        print(f"Hello, my name is {self.name}!")
        
        
  #Call the function  
greet("shreya")

  #Create an instance of the class and call the method  
person = Person("shreya")  
person.greet()

5. What are iterators in Python and how do they differ from iterables?

-> Iterators and iterables are two fundamental concepts in Python that facilitate working with sequences, such as lists, tuples, dictionaries, and sets.

*   Iterables

Iterables are objects that can be iterated over, meaning they can be looped through or have their elements accessed one at a time. Examples of iterables include:

- Lists: [1, 2, 3]  
- Tuples: (1, 2, 3)   
- Dictionaries: {"a": 1, "b": 2}    
- Sets: {1, 2, 3}   
- Strings: "hello"    

Iterables have the __ iter __ method, which returns an iterator object.

* Iterators

Iterators are objects that keep track of their position within an iterable and provide a way to access the elements one at a time.

Iterators have the __ next __ method, which returns the next element from the iterable.

When creating an iterator from an iterable, it starts at the beginning of the iterable. Each time calling the __ next __ method, it returns the next element and moves the iterator's position forward.

An example:


my_list = [1, 2, 3]  # iterable   
my_iter = iter(my_list)  # iterator   

print(next(my_iter))  # prints 1   
print(next(my_iter))  # prints 2    
print(next(my_iter))  # prints 3    



*  Key differences :


*  Creation: An iterable is created when you define a sequence, such as a list or tuple. An iterator is created when you call the iter() function on an iterable.

*  Position tracking: An iterator keeps track of its position within the iterable, while an iterable does not.
*   Element access: An iterator provides access to elements one at a time using the __ next __ method, while an iterable provides access to all elements at once.


*  Statefulness: An iterator is stateful, meaning its position changes after each access. An iterable is stateless, meaning its elements remain unchanged.

Generators are special types of iterators that can be used to generate a sequence of values on-the-fly, instead of storing them all in memory at once. This makes generators particularly useful when working with large datasets or infinite sequences.

*   Defining Generators

Generators are defined using a function with the yield keyword instead of return. When a generator is called, it doesn't execute immediately. Instead, it returns a generator object, which can be iterated over using a for loop or the next() function.

     # Iterate over the generator
for num in gen:

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def numbers(num):
    sum_even = 0
    for i in num:
        if i%2==0:
            sum_even += i

    return sum_even


print(numbers([4,5,9,7,2,8]))
```

4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def numbers(total_nums):
    prime = []

    for n in total_nums:
        if n<=1:
            continue
        is_prime = True
        for i in range(2,int(n**0.5)+1):
            if n%i ==0:
                is_prime = False
                break
        if is_prime:
            prime.append(n)

    return f"the prime numbers from 1 to 200 are : {prime}"

print(numbers(i for i in range(1,201)))
```

5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.i += 1
        return result


fib = FibonacciIterator(8)
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):

    for i in range(exponent + 1):
        yield 2** i

for power in powers_of_two(5):
    print(power)
```

7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(file_path):

    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
        return


file_path = 'example.txt'
for line in read_file_lines(file_path):
    print(line)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
def sorting(tuple_list):
    sorted_tuples = sorted(tuple_list, key=lambda x: x[1])
    return sorted_tuples

print(sorting([(9,7),(4,2),(3,6)]))
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temperatures = [0, 10, 20, 30, 40]
print("Celsius temperatures:-\n",celsius_temperatures)
print("converting Celsius to Fahrenheit:-")
fahrenheit_temperatures = list(map(lambda celsius: (celsius * 9/5) + 32, celsius_temperatures))
print(list(map(round, fahrenheit_temperatures)))
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
input_string = "Hello, World!"
result = "".join(filter(lambda char: char.lower() not in 'aeiou', input_string))

print("Original String:", input_string)
print("String without Vowels:", result)
```
